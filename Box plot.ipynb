```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
plt.style.use('fivethirtyeight')
```

```python
tips.head()
```

```python
sns.boxplot(x = tips['size'])
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x='sex', y='total_bill',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='coolwarm')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time', palette='coolwarm')
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips, order=['Female','Male'])
```

```python
sns.boxplot(data=tips, orient='horizontal', palette='husl')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
sns.swarmplot(x='day', y='total_bill', data=tips,color='0.5')
```
