```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot('year', 'month','passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False, cmap='coolwarm', linewidths=1)
```

```python
sns.clustermap(flights, row_cluster=False, cmap='Blues', linewidths=1)
```

```python
sns.clustermap(flights, row_cluster=False, cmap='Blues', linewidths=1, figsize=(6,4))
```

```python
#standard of columns
sns.clustermap(flights, row_cluster=False, cmap='Blues', linewidths=1, figsize=(6,4), standard_scale=1)
```

```python
#standard of rows
sns.clustermap(flights, row_cluster=False, cmap='Blues', linewidths=1, figsize=(6,4), standard_scale=0)
```

```python
#normalizing columns
sns.clustermap(flights, z_score=1)
```

```python
#normalizing rows
sns.clustermap(flights, z_score=0)
```
