```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(x=tips['tip'], color='green')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=0.3)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips, jitter=1)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,jitter=1, linewidth=1)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips, hue='sex', jitter=1, linewidth=1)
```

```python
sns.stripplot(x='day',y='total_bill', hue='sex',data=tips, jitter=0.1, dodge=True, palette='winter')
```

```python
sns.stripplot(x='day',y='total_bill', size=10,data=tips, alpha=0.4, jitter=0.1, dodge=True, marker='d', palette='winter')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, color='0.5')
sns.boxplot(x='day',y='total_bill',data=tips, palette='winter')
```
