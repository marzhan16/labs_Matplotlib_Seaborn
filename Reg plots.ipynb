```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(x = 'total_bill', y='tip', data=tips)
```

```python
sns.regplot(x = 'total_bill', y='tip', data=tips, color='red')
```

```python
mean=[2,5]
cov=[[1.1,0.4],[2.2,3]]
x_axis,y_axis=np.random.multivariate_normal(mean,cov,100).T
sns.regplot(x=x_axis,y=y_axis)
```

```python
series_x_value = pd.Series(x_axis, name='X-value')
series_y_value = pd.Series(y_axis, name='Y-value')
sns.regplot(x=series_x_value, y= series_y_value, marker='D',color='grey')
```

```python
sns.regplot(x = 'total_bill', y='tip', data=tips,marker='s',\
           scatter_kws={'color':'blue'},\
           line_kws={'color':'red','linewidth':5})
```

```python
sns.regplot(x = 'total_bill', y='tip', data=tips, ci=12)
```

```python
sns.regplot(x = 'size', y='total_bill', data=tips)
```

```python
sns.regplot(x = 'size', y='total_bill', data=tips, x_jitter=0.1)
```

```python
sns.regplot(x = 'size', y='total_bill', data=tips, x_estimator=np.mean())
```
