```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, hue='species', palette='husl', markers=['o','s','+'])
```

```python
sns.pairplot(iris, vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris, vars=['sepal_length','sepal_width'], size=2,aspect=3)
```

```python
sns.pairplot(iris, x_vars=['sepal_length','sepal_width'], y_vars=['petal_length','petal_width'],\
            palette='husl', hue='species')
```

```python
sns.pairplot(iris, diag_kind='kde', palette='winter', hue='species')
```

```python
sns.pairplot(iris, kind='reg', palette='winter', hue='species')
```
