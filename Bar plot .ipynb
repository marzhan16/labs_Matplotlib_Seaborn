```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex', palette='winter_r')
```

```python
sns.barplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.barplot(x='total_bill',y='day',data=tips,palette='spring')
```

```python
sns.barplot(x='day', y='tip',data=tips, palette='spring', order=['Sat', 'Sun','Fri','Thur'])
```

```python
from numpy import median
```

```python
sns.barplot(x='day', y='tip', data=tips, estimator=median)
```

```python
sns.barplot(x='smoker', y='tip', data=tips, estimator=median, hue='sex', palette='coolwarm')
```

```python
sns.barplot(x='smoker', y='tip', data=tips, ci=22)
```

```python
sns.barplot(x='day', y='tip', data=tips, capsize=0.3)
```

```python
sns.barplot(x='size', y='tip', data=tips, capsize=0.5, palette='husl')
```

```python
sns.barplot(x='size', y='tip', data=tips, capsize=0.5, color='green')
```

```python
sns.barplot(x='size', y='tip', data=tips, capsize=0.5, color='green', saturation=0.4)
```
