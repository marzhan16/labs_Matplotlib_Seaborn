```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris)
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris,kind='reg')
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris,kind='reg', color='green')
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris,kind='kde')
```

```python
#stat_func
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='size', y='total_bill', data=tips, stat_func=spearmanr)
```

```python
sns.jointplot(x='size', y='total_bill', data=tips, stat_func=spearmanr, ratio=1, size=4)
```
