```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
plt.style.use('fivethirtyeight')
```

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x= x.map_diag(plt.hist)
```

```python
x = sns.PairGrid(iris)
x= x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue='species')
x= x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue='species', palette='RdBu')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue='species')
x= x.map_diag(plt.hist, histtype='step')
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, hue='species')
x= x.map_diag(plt.hist, histtype='step', linewidth = 3)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(iris, vars = ['petal_length', 'petal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris, vars = ['petal_length', 'petal_width'], hue='species')
x = x.map_offdiag(plt.scatter, edgecolor='black')
x = x.map_diag(plt.hist, edgecolor='black')
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, x_vars=['petal_length','petal_width'], y_vars=['sepal_length','sepal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
x = sns.PairGrid(iris, hue='species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, hue='species', hue_kws={'marker':['D','s','+']})
x = x.map(plt.scatter, s=10)
x = x.add_legend()
```
