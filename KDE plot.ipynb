```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Kernel Dens. Estim PDF
```

```python
mean = [0,0]
cov = [[0.2,0],[0,3]]
```

```python
x, y = np.random.multivariate_normal(mean, cov,size=40).T
```

```python
sns.kdeplot(x)
```

```python
sns.kdeplot(y)
```

```python
sns.kdeplot(y, shade=True, color='red')
```

```python
sns.kdeplot(x,y)
```

```python
sns.kdeplot(x,y, cmap='RdBu')
```

```python
sns.kdeplot(x,y, cmap='RdBu', n_levels=18)
```

```python
sns.kdeplot(x, bw=10)
```

```python
sns.kdeplot(y, bw=0.1)
```

```python
iris = sns.load_dataset('iris')
```

```python
setosa = iris.loc[iris.species=='setosa']
versicolor = iris.loc[iris.species=='versicolor']
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width)
sns.kdeplot(versicolor.sepal_length, versicolor.sepal_width)
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='RdBu')
sns.kdeplot(versicolor.sepal_length, versicolor.sepal_width, cmap='winter')
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='RdBu', shade=True)
sns.kdeplot(versicolor.sepal_length, versicolor.sepal_width, cmap='winter')
```

```python
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='RdBu', n_levels=15)
sns.kdeplot(versicolor.sepal_width, versicolor.sepal_length, cmap='winter')
```
