```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(x = tips['tip'])
```

```python
sns.violinplot(x = tips['size'], y=tips['total_bill'])
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', order=['Sat', 'Thur','Fri','Sun'])
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', order=['Sat', 'Thur','Fri','Sun'], split=True)
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='quartile')
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='quartile',split=True)
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='quartile')
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='stick')
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='stick', scale='count')
```

```python
sns.violinplot(x = 'day', y='total_bill', data=tips, hue='smoker', inner='stick', scale='count', scale_hue=False)
```
