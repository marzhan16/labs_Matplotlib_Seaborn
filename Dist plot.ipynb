```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
num = np.random.randn(150)
```

```python
sns.distplot(num, color='red')
```

```python
label_dist= pd.Series(num, name='variable x')
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist, color='green', vertical= True, hist_kws=dict(edgecolor="k"))
```

```python
sns.distplot(label_dist, color='green', hist=False)
```

```python
sns.distplot(label_dist, color='green', rug = True, hist=False)
```
