```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
normal = np.random.rand(12,15)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal, annot=True)
```

```python
sns.heatmap(normal, vmin=0, vmax=2)
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights=flights.pivot('year','month','passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights, annot=True)
```

```python
sns.heatmap(flights, annot=True, fmt='d')
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidth=0.5)
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidth=0.5, vmin=100,vmax=500)
```

```python
sns.heatmap(flights, annot=True, fmt='d', linewidth=0.5, cmap='RdBu')
```
