```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.FacetGrid(tips, row='smoker', col='time')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill', color='blue', bins=15)
```

```python
x = sns.FacetGrid(tips, row='time', col='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, row='time', col='smoker')
x = x.map(sns.regplot, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
x = x.add_legend()
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'total_bill', 'time')
```

```python
x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', size=4)
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', size=5, aspect=0.3)
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
x = sns.FacetGrid(tips, col='day', size=5, aspect=0.3, col_order=['Fri','Sun','Sat','Thur'])
x = x.map(sns.boxplot, 'time', 'total_bill')
```
